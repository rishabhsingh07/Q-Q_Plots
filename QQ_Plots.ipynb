# It is used to check whether our data is normal/gaussian distribution or not

```python
import numpy as np
import statsmodels.api as sm
import pylab as py
```

#### Lets first create a normal distribution data:

```python
# Standard normal distribution

data_points = np.random.normal(0, 1, 100)
```

#### We can see in the below graph that most of our blue data points are almost on or close to the red line, that shows that the data is normally distributed data

```python
sm.qqplot(data_points, line='45')
py.show()
```

```python
import seaborn as sns
sns.distplot(data_points,kde=False)
```

```python
sns.distplot(data_points)
```

#### Lets create a dataset which is not gaussian/normal distributed

```python
# Log normal data
data = np.random.lognormal(0, 1, 100)
```

#### We can see in the image below that most of our blue data points are not on the red line, thus confirming that the dataset is not normally distributed

```python
sm.qqplot(data, line='45')
py.show()
```

```python
sns.distplot(data)
```

#### Let's log transform our log normal data to normal distribution and check the qq plot again:

```python
sm.qqplot(np.log(data),line='45')
py.show()
```

#### We can see in the image above that our dataset has again been converted to a gaussian/normal distribution, as the datapoints are again on the red line
